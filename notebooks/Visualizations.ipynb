```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataDir = "../data/"
df = pd.read_csv(dataDir + "regression_data.csv")
df.head()
```

```python
df = df[["density", 
         "administrative and support and waste management and remediation services",
         "transportation and warehousing", 
         "wholesale trade",
         "manufacturing"]]
df = df.replace(to_replace=float(0), value=np.nan)
admin = df["administrative and support and waste management and remediation services"]
trans = df["transportation and warehousing"]
whole = df["wholesale trade"]
manu = df["manufacturing"]
```

```python
sns.set(style="darkgrid")
fig = plt.figure(figsize=(10, 10))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
ax = fig.add_subplot(2, 2, 1)
sns.scatterplot(x='density',y="administrative and support and waste management and remediation services", data=df)
plt.hlines(admin.mean(), 0, 20000)
plt.hlines([admin.mean() + admin.std(), admin.mean() - admin.std()], 0, 20000,
           color = 'r', linestyle='--')
ax.set(ylabel='administrative')
ax = fig.add_subplot(2, 2, 2)
sns.scatterplot(x='density',y='transportation and warehousing', data=df)
plt.hlines(trans.mean(), 0, 20000)
plt.hlines([trans.mean() + admin.std(), trans.mean() - trans.std()], 0, 20000,
           color = 'r', linestyle='--')
ax = fig.add_subplot(2, 2, 3)
sns.scatterplot(x='density',y="wholesale trade", data=df)
plt.hlines(whole.mean(), 0, 20000)
plt.hlines([whole.mean() + whole.std(), whole.mean() - whole.std()], 0, 20000,
           color = 'r', linestyle='--')
ax = fig.add_subplot(2, 2, 4)
sns.scatterplot(x='density',y="manufacturing", data=df)
plt.hlines(manu.mean(), 0, 20000)
plt.hlines([manu.mean() + manu.std(), manu.mean() - manu.std()], 0, 20000,
           color = 'r', linestyle='--')
plt.show()
```
